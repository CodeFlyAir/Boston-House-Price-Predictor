```python
from sklearn.model_selection import StratifiedShuffleSplit
split=StratifiedShuffleSplit(n_splits=1,test_size=0.2,random_state=42)
for train_set,test_set in split.split(housing,housing['CHAS']):
    strat_test_set = housing.loc[test_set]
    strat_train_set = housing.loc[train_set]
```

```python
housing_train_data = housing_train.drop('MEDV',axis=1)
housing_train_labels = housing_train['MEDV'].copy()

housing_train_labels.head()
```

```python
housing_test_data = housing_test.drop('MEDV',axis=1)
housing_test_labels = housing_test['MEDV'].copy()

housing_test_labels.head()
```

# Creating Pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer

pipeline = Pipeline([
    ('impute',SimpleImputer(strategy="median")),
     ('std_scaler',StandardScaler())
])
```

```python
housing_train_transform = pipeline.fit_transform(housing_train_data)
```

```python
housing_train_transform.shape
```

# Model

## Prediction

```python
def data_transform():
    return pipeline.transform(housing_train_data)
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
model_lin = LinearRegression()
model_lin.fit(housing_train_transform,housing_train_labels)
predict_data = data_transform()
```

## Evaluating Model

```python
from sklearn.metrics import mean_squared_error
prediction = model_lin.predict(predict_data)
mse = mean_squared_error(housing_train_labels,prediction)
rmse = np.sqrt(mse)
```

```python
rmse
```

# Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
model_dec = DecisionTreeRegressor()
model_dec.fit(housing_train_transform,housing_train_labels)
predict_data = data_transform()
```

## Evaluating Model

```python
from sklearn.metrics import mean_squared_error
prediction = model_dec.predict(predict_data)
mse = mean_squared_error(housing_train_labels,prediction)
rmse = np.sqrt(mse)
rmse
```

# Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
model_rfr = RandomForestRegressor()
model_rfr.fit(housing_train_transform,housing_train_labels)
predict_data = data_transform()
```

## Evaluating Model

```python
from sklearn.metrics import mean_squared_error
prediction = model_rfr.predict(predict_data)
mse = mean_squared_error(housing_train_labels,prediction)
rmse = np.sqrt(mse)
rmse
```

# Cross Validation - For Decision Tree Regressor

```python
from sklearn.model_selection import cross_val_score
score = cross_val_score(model_dec, housing_train_data, housing_train_labels,scoring="neg_mean_squared_error", cv=20)
rmse_score = np.sqrt(-score)
rmse_score
print(f"Mean : {np.mean(rmse_score)}\nStandard Deviation : {np.std(rmse_score)}")
```

# Cross Validation - For Linear Regression

```python
from sklearn.model_selection import cross_val_score
score = cross_val_score(model_lin, housing_train_data, housing_train_labels,scoring="neg_mean_squared_error", cv=20)
rmse_score = np.sqrt(-score)
rmse_score
print(f"Mean : {np.mean(rmse_score)}\nStandard Deviation : {np.std(rmse_score)}")
```

# Cross Validation - For Random Forest Regressor

```python
from sklearn.model_selection import cross_val_score
score = cross_val_score(model_rfr, housing_train_data, housing_train_labels,scoring="neg_mean_squared_error", cv=20)
rmse_score = np.sqrt(-score)
rmse_score
print(f"Mean : {np.mean(rmse_score)}\nStandard Deviation : {np.std(rmse_score)}")
```

## Model Testing

```python
housing_test_transform = pipeline.fit_transform(housing_test_data)
test_predict_data = pipeline.transform(housing_test_data)
lin_predict = model_lin.predict(test_predict_data)
dec_predict = model_dec.predict(test_predict_data)
rfr_predict = model_rfr.predict(test_predict_data)
print(f"RMSE for Linear Regression : {np.sqrt(mean_squared_error(housing_test_labels,lin_predict))}")
print(f"RMSE for Decision Tree Regressor : {np.sqrt(mean_squared_error(housing_test_labels,dec_predict))}")
print(f"RMSE for Random Forest Regressor : {np.sqrt(mean_squared_error(housing_test_labels,rfr_predict))}\n")
```
